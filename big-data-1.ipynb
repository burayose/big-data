```python
#1 In which release_date there was the highest average revenue use ? (by decending order)

df.groupby('release_date')['revenue'].mean().sort_values(ascending=False)
```

```python
#1 which title is a highest average revenue  ? (by decending order)

df.groupby('title')['revenue'].mean().sort_values(ascending=False)
```

```python
#1 which spoken_languages is a hignest average revenue ?
# Group by 'spoken_languages' and 'title', calculate mean revenue, and sort values
result = df.groupby(['spoken_languages', 'title'])['revenue'].mean().sort_values(ascending=False)

# Display the language and movie titles with the highest average revenue
print(result.head(3))  # This will show the row with the highest average revenue
```

```python
# visualize spoken languages and revenue

sns.barplot(x='spoken_languages', y='revenue', data=df)
plt.title("Average revenue by spoken_languages")
plt.show()
```

```python
#2 Display the movie titles with the highest revenue relative to budget
#Group by 'budget' and 'title', calculate mean revenue, and sort values

result = df.groupby(['budget', 'title'])['revenue'].sum().sort_values(ascending=False)

print(result.head(3))  # This will show the row with the highest total revenue related to budget
```

```python
#3 Display the movie titles with the highest revenue relative to release_date ?

# Assuming df is the DataFrame containing the movie data with columns: 'release_date', 'title', 'revenue'

# Group by 'release_date' and 'title', calculate mean revenue, and sort values
result = df.groupby(['release_date', 'title'])['revenue'].sum().sort_values(ascending=False)

# Display the top 10 rows with the highest total revenue related to release_date
print(result.head(10))
```

```python
# visualize release_date and revenue

sns.barplot(x='release_date', y='revenue', data=df)
plt.title("Average revenue by release_date")
plt.show()
```

```python
#4 Group by 'production_companies' and 'title', calculate mean popularity, and sort values
result = df.groupby(['production_companies', 'title'])['popularity'].mean().sort_values(ascending=False)

print(result.head(5))  # This will show the row with the highest mean popularity related to each production company and title
```

```python
# visualize

plt.subplots(figsize=(10, 6.5))

# Scatter plot of runtime
sns.scatterplot(x=[i for i in range(len(df.popularity))], y=df.popularity)

# Line plot for mean and median runtime
sns.lineplot(x=[i for i in range(len(df.popularity))], y=df.popularity.mean(), label="mean")
sns.lineplot(x=[i for i in range(len(df.popularity))], y=df.popularity.median(), label="median")

plt.legend()  # Display legend
plt.show()
```

```python
#5  Group by 'spoken_languages' and 'title', calculate mean popularity, and sort values

result = df.groupby(['spoken_languages', 'title'])['popularity'].mean().sort_values(ascending=False)

print(result.head(5))  # Display the top 5 rows with the highest average popularity related to spoken languages
```

```python

#6 Select top 10 movies based on popularity and display their titles and release dates

top10 = df.nlargest(10, 'popularity')[['title', 'release_date', 'popularity']]
print(top10)

# Plotting using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(top10['release_date'], top10['popularity'])
plt.xlabel('Release Date')
plt.ylabel('Popularity')
plt.title('Top 10 Movies by Popularity and Release Date')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
